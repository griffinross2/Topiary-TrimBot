TPS92365x power supply calculations

First define parameters

```python
Vin = 24
Vf_led = 22.4
Vout = Vf_led + Vin
If_led = 0.018 # 18 mA (little threshold under 20mA)
L_ripple = 1 # 100% of max current
Iout_ripple = 0.001
f_sw = 2200000
```

Calculate parameters

```python
IL_max = (Vin + Vf_led) * If_led / Vin
print(f"Max inductor current: {IL_max*1e3:0.2f} mA")

L = (Vin * Vf_led) / ((Vf_led + Vin) * L_ripple * IL_max * f_sw)
print(f"Inductor value: {L*1e6:0.2f} uH")
```

The inductor will be too big, so we have to accept a smaller value and operation in discontinuous mode (DCM)

```python
L = 22e-6
print(f"Using inductor value: {(L*1e6):0.2f} uH")

IL_peak = ((2*If_led*Vf_led)/(L*f_sw))**0.5
print(f"Peak inductor current (DCM): {(IL_peak*1e3):0.2f} mA")
```
